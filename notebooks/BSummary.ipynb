```python
stopw = spacy.lang.en.stop_words.STOP_WORDS
fpos = {
  "PUNCT",
  "INTJ",
  "X"
}
```

```python
from sortedcontainers import SortedList
from itertools import chain
import re

def geti(word):
  return word.i

def isgch(c):
  return ( c.dep_ in {
    "compound",
    "aux"
  } or "subj" in c.dep_
  and c.pos_ not in fpos
  )
nfs = nlp(coref_text)

w_re = re.compile("[ \w]+")
subjs = dict()
for w in nfs:
  if('subj' in w.dep_):
    fci = SortedList(key=geti)
    ch = [w]
    while(len(ch)):
      c = ch.pop()
      fci.add(c)
      ch.extend((cc for cc in c.children if isgch(cc)))

    s = [str(f) for f in fci if f.pos_ not in fpos]
    s = [sw for sw in s if sw not in stopw and w_re.match(sw)]
    if(len(s)):
      s = ' '.join(s).strip().lower()

      subl = subjs.get(s)
      if(subl is not None):
        subl.append(fci)
      else:
        subjs[s] = [fci]

def t(v):
  for vv in v:
    for vvv in vv:
      yield (vvv.dep_, vvv.pos_)
print(*((k, list(t(v))) for k,v in subjs.items()), sep='\n')
```

```python
rels = {}
roots = {}
for sstr, s_l in subjs.items():
  for tgr in s_l:
    for tok in tgr:
      ancs = list(tok.ancestors)
      root = None
      while(len(ancs)):
        an = ancs.pop(0)
        if(an.dep_ == "ROOT"):
          root = an
          ancs.clear()
          break
        ancs.extend(an.ancestors)
      if(root is not None):
        rl = SortedList(key=geti)
        q = [root]
        while(len(q)):
          tr = q.pop()
          rl.add(tr)
          q.extend((c for c in tr.children if c.dep_ == "aux"))

        sr = ' '.join((rr.lemma_ for rr in rl))
        if(sstr not in rels):
          rels[sstr] = {sr: [(tgr, rl)]}
        elif(sr not in rels[sstr]):
          rels[sstr][sr] = [(tgr, rl)]
        else:
          rels[sstr][sr].append((tgr, rl))

        if(sr not in roots):
          roots[sr] = [rl]
        else:
          roots[sr].append(rl)

        print(f"{sstr} -> {sr}")
        break
      ancs.clear()
    else:
        print('-')
```

```python

def add_dict(original, added):
  qd = [(original, added)]
  while(len(qd)):
    b, a = qd.pop(0)
    for k, v in a.items():
      if(type(v) == dict):
        d = b.get(k)
        if(d is None):
          b[k] = v
        else:
          qd.append((d, v))
      else:
        b[k] = v

def handle_nsis(ssubj, sl, sroot, rl, obj):
  objqs = []
  tmpq = []
  for tok in obj:
    if(tok.pos_ == "PUNCT"):
      if(len(tmpq)):
        objqs.append(tmpq.copy())
        tmpq.clear()
    else:
      tmpq.append(tok)
  if(len(tmpq)):
    objqs.append(tmpq)

  #print(objqs)

  res = {}
  for sobjq in (' '.join(map(str, oq)) for oq in objqs):
    print(f"{ssubj} -> {sroot} -> {sobjq}")
    objq = objqs.pop(0)
    objqs.append((sl, rl, objq))

  return {ssubj:{sroot:objqs}}

def special_is(ssubj, sl, sroot, rl, obj):
  print("special is")
  hdld = handle_nsis(ssubj, sl, sroot, rl, obj)
  
  for (_, __, spobj) in hdld[ssubj][sroot]:
    nssubj = ' '.join(map(str, spobj))
    print(f"{nssubj} -> {sroot} -> {ssubj}")
    d0 = hdld.get(nssubj)
    if(d0 is None):
      d0 = hdld[nssubj] = dict()
    d1 = d0.get(sroot)
    if(d1 is None):
      d1 = d0[sroot] = []
    d1.append((spobj, rl, sl))

    nadj = [t for t in spobj if t.pos_ != "ADJ"]
    padj = [t for t in spobj if t.pos_ == "ADJ"]

    snadj = None
    if(len(nadj) and len(nadj) != len(spobj)):
      snadj = ' '.join(map(str, nadj))
      print(f"{ssubj} -> {sroot} -> {snadj}")
      print(f"{snadj} -> {sroot} -> {ssubj}") 
      d0 = hdld.get(ssubj)
      if(d0 is None):
        d0 = hdld[ssubj] = dict()
      d1 = d0.get(sroot)
      if(d1 is None):
        d1 = d0[sroot] = []
      d1.append((sl, rl, nadj))

      d0 = hdld.get(snadj)
      if(d0 is None):
        d0 = hdld[snadj] = dict()
      d1 = d0.get(sroot)
      if(d1 is None):
        d1 = d0[sroot] = []
      d1.append((nadj, rl, sl))

    if(len(padj) != len(spobj)):
      for adj in padj:
        sadj = str(adj)
        adj = [adj]
        print(f"{ssubj} -> {sroot} -> {sadj}")
 
        d0 = hdld.get(ssubj)
        if(d0 is None):
          d0 = hdld[ssubj] = dict()
        d1 = d0.get(sroot)
        if(d1 is None):
          d1 = d0[sroot] = []
        d1.append((sl, rl, adj))
 
        if(snadj is not None):
          print(f"{snadj} -> {sroot} -> {sadj}")

          d0 = hdld.get(snadj)
          if(d0 is None):
            d0 = hdld[snadj] = dict()
          d1 = d0.get(sroot)
          if(d1 is None):
            d1 = d0[sroot] = []
          d1.append((nadj, rl, adj))


  return hdld


frel = dict()
for ssubj, rrel in rels.items():
  ssrel = frel[ssubj] = dict()
  for sroot, rls in rrel.items():
    srrel = ssrel[sroot] = []
    for (sl, rl) in rls:
      foundw = set((*sl, *rl))
      obj = SortedList(key=geti)
      ri = next((r.i for r in rl if r.dep_ == "ROOT"))
      q = list((cc for c in rl for cc in c.children if cc.i > ri and cc not in foundw))
      while(len(q)):
        nobj = q.pop()
        foundw.add(nobj)
        obj.add(nobj)
        q.extend((cc for cc in nobj.children if cc.i > ri and cc not in foundw))

      sobj = ' '.join(map(str, obj))
      if(len(obj)):
        npobj = list((o for o in obj if o.pos_ != "PART"))
        if(sroot == 'be' and len(npobj) and npobj[0].pos_ not in {"VERB", "ADP"}):
          add_dict(frel, special_is(ssubj, sl, sroot, rl, obj))
        else:
          add_dict(frel, handle_nsis(ssubj, sl, sroot, rl, obj))
      else:
        srrel.append((sl, rl, obj))
        print(f"{ssubj} -> {sroot} -> {sobj}")
```

```python
stopw = spacy.lang.en.stop_words.STOP_WORDS
v_pos = {
  "NOUN",
  "VERB",
  "ADJ",
  "ADV"
}
```

```python
from itertools import chain
from sortedcontainers import SortedList

check = nlp("a cork strength material in construction")

def get0(t): return t[0]
def get1(t): return t[1]

qss = []
for wn1, w1 in enumerate((w for w in check if str(w) not in stopw)):
    print(f"{str(w1)}: {w1.pos_}")
    if(w1.pos_ in v_pos):
        ss1 = wn.synsets(str(w1), pos=getattr(wn, w1.pos_))
    else:
        ss1 = wn.synsets(str(w1))

    definitions = list(set((
        *ss1, 
        *[s for s1 in ss1 for l in s1.lemmas() if l.name() != s1.name().split('.', 1)[0] for s in wn.synsets(l.name())],
        *[h for s in ss1 for h in s.hypernyms()]
    )))

    computed = set()
    dl = SortedList(key=get1)
    # For each definition
    for sn1, s1 in enumerate(definitions):
        s1name = s1.name()
        if(s1name in computed):
            continue
        computed.add(s1name)

        s1d = nlp(s1.definition())
        qsim = qd.similarity(s1d)

        for ex1 in s1.examples():
            sim = qd.similarity(nlp(ex1))
            if(sim > qsim):
                qsim = sim

        wsim = check.similarity(s1d)

        for ex1 in s1.examples():
            sim = check.similarity(nlp(ex1))
            if(sim > wsim):
                wsim = sim

        
        pdist = 0
        wdist = 0
        dsim = 0

        #print(f" #\n {s1.definition()}\n  base: {qsim:.4f} + {wsim:.4f}", flush=False)
        # Iterate over every word
        for wn2, w2 in enumerate(qd):
            if(wn1 == wn2):
                continue
            #print(f"  In: {w2}\n   Definitions:", flush=False)
            if(w2.pos_ in v_pos):
                ss2 = wn.synsets(str(w2), pos=getattr(wn, w2.pos_))
            else:
                ss2 = wn.synsets(str(w2))

            # For each word definition, the most matching definition
            for s2 in ss2:
                if(s1 == s2):
                    continue
                sim = s1d.similarity(nlp(s2.definition()))
                if(sim > dsim):
                    dsim = sim
                #print(f"    {s2.definition()}\n     {sim:.4f}", flush=False)

            #print("   Examples:", flush=False)
            for s2 in ss2:
                if(s1 == s2):
                    continue
                for ex2 in s2.examples():
                    sim = s1d.similarity(nlp(ex2))
                    if(sim > dsim):
                        dsim = sim
                    #print(f"    {ex2}\n     {sim:.4f}", flush=False)

            #print("   Distances:", flush=False)
            for s2 in ss2:
                if(s1 == s2):
                    continue
                pd = s1.path_similarity(s2)
                if(pd > pdist):
                    pdist = pd
                wd = s1.wup_similarity(s2)
                if(wd > wdist):
                    wdist = wd
                #print(f"    {s2.name().split('.', 1)[0]}: {pd:.4f} | {wd:.4f}", flush=False)

        ssim = (qsim + dsim + wsim) * (pdist + wdist)
        dl.add((s1, ssim))
                
    qss.append(dl)

print("###", flush=False)
for qsl in qss:
    qs = qsl[-1][0]
    print(f"{qs.name().split('.', 1)[0]}: {qs.definition()}", flush=False)
print("###")
```
